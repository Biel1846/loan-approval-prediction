```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load dataset
df = pd.read_csv('data/loan_approval_dataset.csv')

# Cek 5 data teratas
df.head()
```

```python
df.info()
```

```python
# 1. Membersihkan nama kolom (menghilangkan spasi di awal/akhir)
df.columns = df.columns.str.strip()
```

```python
# 2. Menghapus kolom loan_id (tidak kepakai)
df = df.drop(columns=['loan_id'])
```

```python
# 4. Membersihkan spasi pada data string (karena di csv aslinya ada spasi: ' Graduate')
categorical_cols = ['education', 'self_employed', 'loan_status']
for col in categorical_cols:
    df[col] = df[col].str.strip()
```

```python
# Cek nilai unik untuk memastikan sudah bersih
print(df['education'].unique())
print(df['loan_status'].unique())
```

```python
# Lihat sebaran data Target (Approved vs Rejected)
plt.figure(figsize=(6, 4))
sns.countplot(x='loan_status', data=df)
plt.title('Distribusi Status Pinjaman')
plt.show()
```

```python
# Lihat hubungan CIBIL Score dengan Loan Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='loan_status', y='cibil_score', data=df)
plt.title('Hubungan CIBIL Score vs Status Pinjaman')
plt.show()
# Insight: Biasanya CIBIL score tinggi lebih mudah di-approve
```

```python
# Education: Graduate -> 1, Not Graduate -> 0
df['education'] = df['education'].map({'Graduate': 1, 'Not Graduate': 0})
```

```python
# Self Employed: Yes -> 1, No -> 0
df['self_employed'] = df['self_employed'].map({'Yes': 1, 'No': 0})
```

```python
# Loan Status (Target): Approved -> 1, Rejected -> 0
# Kita simpan di variabel y, jadi boleh tetap teks atau ubah angka. 
# Di sini kita biarkan y tetap teks supaya report-nya enak dibaca, atau ubah angka juga boleh.
# Mari kita pisahkan X (fitur) dan y (target)
```

```python
X = df.drop(columns=['loan_status'])
y = df['loan_status']
```

```python
# Cek data hasil encoding
X.head()
```

```python
# Bagi data jadi Training (70%) dan Testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Buat model Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Latih model
clf.fit(X_train, y_train)
```

```python
# Prediksi ke data test
y_pred = clf.predict(X_test)
```

```python
# 1. Cek Akurasi
acc = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {acc * 100:.2f}%")
```

```python
# 2. Classification Report (Detail Precision, Recall, F1-Score)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# 3. Confusion Matrix (Biar tau salah tebaknya dimana)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Approved', 'Rejected'], yticklabels=['Approved', 'Rejected'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
feature_scores = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_scores, y=feature_scores.index)
plt.title("Faktor Paling Penting dalam Persetujuan Pinjaman")
plt.show()
```

```python
import joblib

# Simpan model yang sudah dilatih
joblib.dump(clf, 'loan_model.sav')
print("Model berhasil disimpan!")
```
